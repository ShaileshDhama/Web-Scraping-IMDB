```python
url = 'https://www.imdb.com/chart/boxoffice'
html = requests.get(url)
```

```python
html.text
```

```python
bsObj = BeautifulSoup(html.text, 'lxml')
```

```python
bsObj
```

```python
bsObj.find('title').getText()
```

```python
h1List = bsObj.findAll('h1')
```

```python
for h1 in h1List:
    print(h1)
```

```python
h4List = bsObj.findAll('h4')
```

```python
for h4 in h4List:
    print(h4.getText())
```

```python
titleList = bsObj.findAll('td',{'class':'titleColumn'}) #({'id':''})
```

```python
for title in titleList:
    print(title.getText())
```

# WEB SCRAPING--MINI PROJECT

Count total number of views for all videos on 1st page of Youtube for given keyword
Keyword:Python

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
result = requests.get('https://www.youtube.com/results?search_query=Python')
```

```python
result
```

```python
result.text
```

```python
<div class="yt-lockup-meta "><ul class="yt-lockup-meta-info"><li>6 months ago</li><li>4,022,091 views</li></ul></div>
```

```python
bs = BeautifulSoup(result.text, 'lxml')
```

```python
elements = bs.find_all('ul',class_='yt-lockup-meta-info')
```

```python
len(elements)
```

```python
for ele in elements:
    print(ele)
```

```python
for ele in elements:
    lis = ele.findChildren()
    for li in lis:
        if li.string.endswith('views'):
            print(li.text)
```

```python
totalView = 0
for ele in elements:
    lis = ele.findChildren()
    for li in lis:
        if li.string.endswith('views'):
            temp = li.text
            temp = temp.replace(' views','')
            temp = temp.replace(',','')
            totalView=totalView + int(temp)
            print(totalView)
```
